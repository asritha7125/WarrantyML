```python
segments_df = pd.read_csv("customer_segments.csv")
```

```python
# Replace spaces with underscores and strip whitespaces
warranty_final.columns = [col.strip().replace(" ", "_") for col in warranty_final.columns]
predictions_df.columns = [col.strip().replace(" ", "_") for col in predictions_df.columns]
segments_df.columns = [col.strip().replace(" ", "_") for col in segments_df.columns]
```

```python
# Convert to datetime and format as date
warranty_final['Purchase_Date'] = pd.to_datetime(warranty_final['Purchase_Date']).dt.date
warranty_final['Claim_Date'] = pd.to_datetime(warranty_final['Claim_Date']).dt.date
```

```python
# Export the cleaned data to CSVs
warranty_final.to_csv("warranty_final.csv", index=False)
predictions_df.to_csv("predictions.csv", index=False)
segments_df.to_csv("customer_segments.csv", index=False)
```
